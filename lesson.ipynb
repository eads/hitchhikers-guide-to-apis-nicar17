# Make a request

We need the Python library called Requests. Requests basically a web browser in code. So let's import it:

```python
import requests
```

Boom. Ok, so if this thing is like a web browser, one thing I have to assume is I can get a URL with it. We do that with `requests.get('<url>')`.

```python
response = requests.get('http://npr.org')
```

Now we have a variable called `response`. It's a Python object that represents everything about the request-response cycle involved in pulling up npr.org.

You can mostly ignore most of the request-response cycle because what you really care about is the content. 

```python
response.content
```

```python
response.headers
```
